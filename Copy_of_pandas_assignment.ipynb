**Question 7:** How many customers have profession "Structural Engineer"?**

```python

structural_engineers = data[data['profession'] == 'Structural Engineer']
print(f"Number of customers with profession 'Structural Engineer': {len(structural_engineers)}")
```

**Question 8**. How many male customers are 'Structural Engineer'?**

```python


male_structural_engineers = data[(data['profession'] == 'Structural Engineer') & (data['gender'] == 'Male')]
print(f"Number of male customers who are 'Structural Engineer': {len(male_structural_engineers)}")
```

**Question 9**. Find out the female Structural Engineers from province Alberta (AB)?**

```python


female_structural_engineers_alberta = data[(data['profession'] == 'Structural Engineer') & (data['gender'] == 'Female') & (data['province'] == 'AB')]
print(f"Number of female Structural Engineers from Alberta: {len(female_structural_engineers_alberta)}")
```

**Question 14.** We need to send new cards to the customers well before the expire, how many cards are expiring in 2019?**

```python


# Using sum()
expiring_in_2019_sum = sum(data['cc_exp'].astype(str).str.contains('2019'))
print(f"Number of cards expiring in 2019 (using sum()): {expiring_in_2019_sum}")

# Using count()
expiring_in_2019_count = data[data['cc_exp'].astype(str).str.contains('2019')]['cc_exp'].count()
print(f"Number of cards expiring in 2019 (using count()): {expiring_in_2019_count}")
```

Question 15. How many people use Visa as their Credit Card Provider?**

**Question 16.** Can you find the customer who spent 100 CAD using Visa?**

**Question 18**. Can you tell the top 5 most popular email providers? (e.g. gmail.com, yahoo.com, etc...)**

```python

data['email_domain'] = data['email'].str.split('@').str[1]

# Count the occurrences of each email domain and get the top 5
top_5_email_providers = data['email_domain'].value_counts().head(5)

print("Top 5 most popular email providers:")
top_5_email_providers
```

**Question 19** Is there any customer who is using email with "am.edu"?

```python

am_edu_emails = data[data['email'].apply(lambda x: x.split('@')[1] == 'am.edu')]
print("Customers using email with 'am.edu':")
am_edu_emails
```
